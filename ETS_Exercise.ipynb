## ETS Decomposition of Airline Passengers Dataset:

## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

## Read dataset

```python

Traffic= pd.read_csv('/content/traffic (1).csv',
                       index_col ='DateTime',
                       parse_dates = True)

```

## show dataset

```python
Traffic.head()
```

## Apply ETS

```python
result = seasonal_decompose(Traffic['Vehicles'],
                            model='multiplicative',
                            period=12)

```

## Plot ETS

```python
result.plot()
```
